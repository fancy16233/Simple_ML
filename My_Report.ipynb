```python
import glob

path =r'C:\Users\fancy16233\Desktop\BIGDATA_REPORT'
allFiles = glob.glob(path+"/*.csv")

frame = pd.DataFrame()
list_ = []

for file_ in allFiles:
    df = pd.read_csv(file_,index_col=None, header=0)
    list_.append(df)
    frame = pd.concat(list_)
    
    
```

```python
frame=frame[data.columns.values]
```

```python
frame = frame.fillna(0)
```

```python
frame = frame.ix[:,2:65]
```

```python
target  = frame.ix[:,4]
train  = frame.drop("FTR",1)
```

```python
target
```

```python
train
```

```python
from sklearn import preprocessing
t_le = preprocessing.LabelEncoder()
t_le.fit(target)


target = t_le.transform(target)
print target
```

```python
t_le.inverse_transform(target)
```

```python
list_en = ["Home_en","Away_en","Referee","HTR"]

def labelencoder(target,encoder):
        encoder = preprocessing.LabelEncoder()
        encoder.fit(target)
        return encoder
    
```

```python
list_en[0] = labelencoder(train.HomeTeam,list_en[0])
list_en[1] = labelencoder(train.AwayTeam,list_en[1])
list_en[2] = labelencoder(train.Referee,list_en[2])
list_en[3] = labelencoder(train.HTR,list_en[3])
```

```python
train.HomeTeam = list_en[0].transform(train.HomeTeam)
train.AwayTeam = list_en[1].transform(train.AwayTeam)
train.Referee = list_en[2].transform(train.Referee)
train.HTR = list_en[3].transform(train.HTR)
```

```python
train
```

```python
from sklearn.cross_validation import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train, target, test_size=0.2, random_state=66)
```

```python
X_train
```

```python
y_train
```

```python

from sklearn.svm import SVC
from sklearn import grid_search
parameters = {'kernel':('linear', 'rbf','poly'), 'C':[1, 10,100,1000],'gamma':[1,10,100,1000]}

svr = SVC()
clf = grid_search.GridSearchCV(svr, parameters)

clf.fit(X_train,y_train)
```

```python
clf.best_estimator_
```

```python
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC

classif = OneVsRestClassifier(SVC(C=1,gamma=1,kernel='linear'))

```

```python
classif.fit(X_train,y_train)
```

```python
pred = classif.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test,pred)
```
